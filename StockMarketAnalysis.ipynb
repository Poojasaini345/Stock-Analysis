```python
# Task 1: Display the Read the DataFrame.

import pandas as pd
copy_path="NSE_FO_Project_Excel.csv"
df=pd.read_csv(copy_path)
print(df)
```

```python
# Task 2: Check some basic information

print('Task 2: Display some basic information about data')
print("\n")
print(df.info())
```

```python
# Task 3: Check the missing value in dataframe

print('Task 3: Display the missing value in dataframe')
print("\n")
print(df.isnull().sum())
```

```python
# Task 4: Display summary statisticas of numerical column.

print('Task 4: Display summary statisticas of numerical column.')
print("\n")
print(df.describe())
```

```python
# Task 5: Display calulate the sum and avarge price of Closing Price.

print("Task 5:")
print("Calculate the sum of closing price : ", df['Closing Price'].sum())
print("Calculate the average closing price : ", df['Closing Price'].mean())
```

```python
# Task 6: finding the highest closing price (with using idxmax) and it's crossponding datials.

highest_closing_price = df.loc[df['Closing Price'].idxmax()]
print("Task 6: Highest Closing Price ")
print("\n")
print(highest_closing_price)
```

```python
# Task 7:  line chart for Closing Price over time[Closing Price or Reporting Date]

print("Task 8: Plot a line chart for Closing Price over time")

import matplotlib.pyplot as plt
df[' Reporting Date'] = pd.to_datetime(df[' Reporting Date'])
plt.plot(df[' Reporting Date'], df['Closing Price'])

# plt.grid(True)
plt.xlabel(' Reporting Date')
plt.ylabel('Closing Price')
plt.title('Closing Price Over Time')
plt.show()
```

```python
# Task 8 : Time Series Plot for Total Trading Volume


plt.figure(figsize=(12, 6))
plt.plot(df[' Reporting Date'], df['Total Trading Volume'], label='Total Trading Volume', marker='o', color='orange')
plt.title('Time Series Plot for Total Trading Volume')
plt.xlabel(' Reporting Date')
plt.ylabel('Total Trading Volume')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# Task 9: Plot for Opening Price vs Closing Price

top5_df = df.head(5)

# Scatter plot for Opening Price vs Closing Price using the top 5 data
plt.figure(figsize=(10, 6))
plt.plot(top5_df['Opening Price'], top5_df['Closing Price'])
plt.title('Opening Price vs Closing Price (Top 5 Data)')
plt.xlabel('Opening Price')
plt.ylabel('Closing Price')
plt.show()
```

```python
# Task 10: Histogram Plot for Strike price.


plt.hist(df['Strike Price '], bins=20, color='red', alpha=0.5, label='Strike Price')

# Add labels and title
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

```python
# Task 11: Calculate the total trading volume .

print("Task 9: Calculate the total trading volume")

total_trading_volume = df['Total Trading Volume'].sum()
print(f"Total Trading Volume: {total_trading_volume}")
```

```python
# Task 12 : Display basic statistics for 'Total Trading Volume' and 'Total Transfer with Option Type'. 

descriptive_stats = df.groupby('Option Type')[['Total Trading Volume', 'Total Transfer Value']].describe()
print(descriptive_stats)
```

```python
# Task 13 : Correlation between 'Option Type', 'Total Trading Volume', 'Total Transfer Value'.



correlation_matrix = df.groupby('Option Type')[['Total Trading Volume', 'Total Transfer Value']].corr()
print(correlation_matrix)

```

```python
# Task 14 : Scatter plot to the relationship between 'Total Trading Volume, and 'Total Transfer Value' for each 'Option Type'.

plt.figure(figsize=(10, 6))
for option_type, data in df.groupby('Option Type'):
    plt.scatter(data['Total Trading Volume'], data['Total Transfer Value'], label=option_type, alpha=0.7)

plt.title('Scatter Plot: Total Trading Volume vs. Total Transfer Value')
plt.xlabel('Total Trading Volume')
plt.ylabel('Total Transfer Value')
plt.legend()
plt.show()
```

```python
# Task 15 : Distribution Plot for Option Types with using Bar Chart

plt.figure(figsize=(8, 5))
df['Option Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Option Types')
plt.xlabel('Option Type')
plt.ylabel('Count')
plt.show()

print('PE (Put Option)')
print('CE (Call Option)')
```

```python
# Task 16 : Box plot for Strike Price vs Option Type

plt.figure(figsize=(10, 6))
df.boxplot(column='Strike Price ', by='Option Type')
plt.title('Box Plot of Strike Price by Option Type')
plt.suptitle('')  # To remove default title from pandas boxplot
plt.show()

```

```python
# Task 17 : Correlation Matrix with Closing Price, Total Trading Volume, Strike Price and Open Interest.

import numpy as np

correlation_matrix = df[['Closing Price', 'Total Trading Volume', 'Strike Price ', 'Open Interest']].corr()
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix )
plt.colorbar()
plt.xticks(np.arange(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(np.arange(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```
